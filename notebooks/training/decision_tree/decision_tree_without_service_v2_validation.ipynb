## Decision Tree - Datensatz ohne`service`- mit Validierungsmenge
- Aufteilen in Trainings-, Validierungs- und Testmengen
- Unser Datensatz hat über 15.000 Einträge, daher macht es Sinn, die Daten in Trainings-, Validierungs- und Testmengen aufzuteilen.
- Wir teilen die Daten wie folgt auf: 60% für das Training, 20% für die Validierung und 20% für den Test.

#### Vorgehen: 

1. **Datenaufteilung**: Die gesamte Datenmenge wird in Trainings-, Validierungs- und Testmengen aufgeteilt.
2. **Hyperparameter-Tuning**: GridSearchCV wird unter Verwendung der Trainingsdaten durchgeführt, wobei es intern eine Cross-Validation verwendet.
3. **Validierung**: Das beste Tuning-Modell wird auf den Validierungsdaten evaluiert.
4. **Abschlussbewertung**: Das finale Modell wird auf den Testdaten bewertet, um die Generalisierungsfähigkeit des Modells zu bestimmen.

```python
import pandas as pd
import geopandas as gpd

from sklearn.model_selection import train_test_split, GridSearchCV, KFold, cross_val_score, cross_val_predict, learning_curve
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, median_absolute_error
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# GeoJSON-Datei laden
gdf = gpd.read_file('../../../data/processed_data/simra_osm_no_service_all.geojson')
```

```python
df = gdf.drop(columns=['markers', 'id', 'index_right', 'geometry', 'highway', 'incidents', 'rides', 'type'])
df.head(2)
```

```python
# Features und Zielvariable trennen
X = df.drop(columns='score')
y = df['score']
```

### 1. Aufteilen in Trainings-, Validierungs- und Testmengen (60% Training, 20% Validation, 20% Test)

```python
# Daten aufteilen: 80% Training + Validierung, 20% Test
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Weiter aufteilen: 80% Training, 20% Validierung
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.2, random_state=1)
```

1. Zuerst werden die Daten in Trainings- und Validierungsdaten (80%) und Testdaten (20%) aufgeteilt.
2. Anschließend wird der Trainings- und Validierungsdatensatz nochmals aufgeteilt in Trainingsdaten (80%) und Validierungsdaten (20%).

Zusammen ergibt sich dadurch eine Aufteilung von etwa 60% Trainingsdaten, 20% Validierungsdaten und 20% Testdaten.

Warum das sinnvoll ist:

- **Trainingsmenge**: Zum Trainieren des Modells.
- **Validierungsmenge**: Wird während des Trainings verwendet, um die Hyperparameter zu optimieren (z.B. über GridSearchCV). Die Validierung hilft sicherzustellen, dass das Modell nicht überangepasst wird und gut auf "unbekannte" Daten generalisiert.
- **Testmenge**: Diese Daten werden erst nach dem gesamten Trainingsprozess verwendet, um die endgültige Leistung des Modells zu beurteilen. Das Modell sieht diese Daten nicht während des Trainingsprozesses.

```python
print(f"Anzahl der Trainingsdaten: {X_train.shape[0]}")
print(f"Anzahl der Validierungsdaten: {X_val.shape[0]}")
print(f"Anzahl der Testdaten: {X_test.shape[0]}")
```

### 2. Hyperparameter-Tuning mit GridSearchCV

```python
# DecisionTreeRegressor-Modell erstellen
dt = DecisionTreeRegressor(random_state=1)
```

```python
# Hyperparameter-Tuning mit GridSearchCV
param_grid = {
    'max_depth': range(1, 21), 
    'min_samples_split': [2, 10, 20], 
    'min_samples_leaf': [1, 5, 10], 
    'max_features': [None, 0.5, 'sqrt', 'log2']
}
```

#### GridSearch & 5-fache Cross Validation

```python
# GridSearchCV einrichten (5-fache Cross-Validation auf Trainingsmenge)
grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, scoring='r2', cv=5, n_jobs=-1)
```

```python
# Modelltraining mit GridSearchCV auf der Trainingsmenge
grid_search.fit(X_train, y_train)
```

```python
# Beste Parameter und Modell
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_

print(f"Best Score: {grid_search.best_score_}")
print(f"Best Params: {best_params}")
```

```python
# Prüfung auf Validierungsdaten
y_val_pred = best_model.predict(X_val)

val_mse = mean_squared_error(y_val, y_val_pred)
val_r2 = r2_score(y_val, y_val_pred)
val_mae = mean_absolute_error(y_val, y_val_pred)
val_medae = median_absolute_error(y_val, y_val_pred)
val_rmse = np.sqrt(val_mse)

print("\nBewertung auf den Validierungsdaten:")
print(f'Mean Squared Error: {val_mse}')
print(f'R²-Score: {val_r2}')
print(f'Mean Absolute Error (MAE): {val_mae}')
print(f'Median Absolute Error: {val_medae}')
print(f'Root Mean Squared Error (RMSE): {val_rmse}')
```

```python
# Abschlussbewertung auf den Testdaten
y_test_pred = best_model.predict(X_test)

test_mse = mean_squared_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
test_medae = median_absolute_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)

print("\nBewertung auf den Testdaten:")
print(f'Mean Squared Error: {test_mse}')
print(f'R²-Score: {test_r2}')
print(f'Mean Absolute Error (MAE): {test_mae}')
print(f'Median Absolute Error: {test_medae}')
print(f'Root Mean Squared Error (RMSE): {test_rmse}')
```

## Fehlermetriken:

**MSE (Mean Squared Error)**:
        Validierungsmenge: 0.0017
        Testmenge: 0.0020
        Der MSE gibt an, wie weit die vorhergesagten Werte im Durchschnitt von den tatsächlichen Werten abweichen. Kleinere Werte sind besser, aber der Unterschied zwischen den beiden Mengen ist gering.

**RMSE (Root Mean Squared Error)**:
        Validierungsmenge: 0.0411
        Testmenge: 0.0445
        RMSE ist die Quadratwurzel des MSE. Diese Werte zeigen, dass die Fehler in beiden Datensätzen ähnlich sind und sich in einem Bereich von ca. 0.041 bis 0.045 befinden.

**MAE (Mean Absolute Error)**:
        Validierungsmenge: 0.0156
        Testmenge: 0.0151
        Der MAE gibt den durchschnittlichen absoluten Fehler an. Auch hier sind die Werte relativ klein, was gut ist. Allerdings zeigt dies nicht, ob das Modell systematisch schlechte Vorhersagen trifft.

**R² (Bestimmtheitsmaß)**:
        Validierungsmenge: -0.0169
        Testmenge: -0.0288
        Das R²-Maß gibt an, wie gut das Modell die Varianz der Zielvariablen erklärt. Ein negativer Wert bedeutet, dass das Modell schlechter ist als ein einfaches Mittelwertmodell (das heißt, das Modell ist im Grunde unbrauchbar, da es die Daten nicht besser als ein Mittelwert vorhersagt).

**Interpretation**:

- MSE, RMSE, und MAE sind relativ niedrig, was auf geringe Vorhersagefehler hinweist, aber diese Werte müssen in den Kontext der Daten (Größenordnung der Zielwerte) gesetzt werden.
- Der negative R²-Wert ist problematisch. Er deutet darauf hin, dass das Modell auf den Validierungs- und Testmengen schlecht generalisiert und im Grunde nicht viel Varianz in den Daten erklärt. Ein negativer R² bedeutet, dass das Modell schlechter ist als eine einfache Mittelwertvorhersage.

### Learning Curve (Lernkurve) 
- ist ein Diagramm, das die Leistung eines Modells sowohl auf den Trainingsdaten als auch auf den Validierungsdaten in Abhängigkeit von der Anzahl der Trainingsdaten zeigt.

Was zeigt eine Learning Curve?
1. Training Score: Die Modellleistung (R²-Score) auf den Daten, auf denen das Modell trainiert wurde.
2. Validation Score: Die Modellleistung auf einem separaten Validierungsdatensatz, der während des Trainings nicht verwendet wurde.

```python
# Learning Curve
train_sizes, train_scores, val_scores = learning_curve(
    best_model, X_train, y_train, cv=5, scoring='r2', n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10)
)

train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
val_scores_mean = np.mean(val_scores, axis=1)
val_scores_std = np.std(val_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color="r")
plt.fill_between(train_sizes, val_scores_mean - val_scores_std, val_scores_mean + val_scores_std, alpha=0.1, color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Trainings Score')
plt.plot(train_sizes, val_scores_mean, 'o-', color='g', label='Validierungs Score')

plt.xlabel('Trainingsmenge')
plt.ylabel('Score (R²)')
plt.title('Lernkurven')
plt.legend(loc='best')
plt.grid()
```

#### Interpretation:

1. Anfangsphase (wenig Trainingsdaten):
   - Trainingskurve hoch, Validierungskurve niedrig: Dein Modell passt sich gut den wenigen Trainingsdaten an und erzielt hohe Werte auf diesen Daten.
   - Validierungskurve niedrig: Das Modell generalisiert schlecht, da es nur wenige Daten gesehen hat und möglicherweise stark überanpasst.
2. Mit zunehmender Trainingsmenge:
   - Anpassung verbessert sich: Das Modell hat mehr Daten gesehen und kann besser generalisiert werden.
   - Trainingskurve sinkt: Die Leistungsfähigkeit auf den Trainingsdaten könnte leicht abnehmen, da das Modell mehr Daten und Variationen erfassen muss (weniger Overfitting).
   - Validierungskurve steigt: Die Leistung auf den Validierungsdaten verbessert sich, da das Modell besser generalisieren kann und mehr Muster versteht.


## Mögliche Problem und nächste Schritte: 
1. **Overfitting oder Underfitting**: Das Modell könnte entweder zu stark auf die Trainingsdaten angepasst sein (Overfitting) oder zu einfach sein, um die zugrunde liegende Komplexität der Daten zu erfassen (Underfitting).
--> Die max_depth des Entscheidungsbaums berprüfen. Eine zu große oder zu kleine Tiefe könnte zu diesen Ergebnissen führen.

2. **Mehr Feature Engineering**: Eventuell könnte das Modell von zusätzlichen Merkmalen oder einer besseren Verarbeitung der vorhandenen Features profitieren.

3. **Größere Hyperparameter-Suche**: Ein breiteres Spektrum an Hyperparametern verwenden, um herauszufinden, ob das Modell verbessert werden kann.

4. **Alternative Modelle**: Versuchen andere Modelle wie Random Forest oder Gradient Boosting zu verwenden, die oft bessere Generalisierungsergebnisse liefern.

5. **Normalisierung der Daten**: Prüfen, ob eine Skalierung der Eingangsfeatures hilfreich ist, besonders wenn die Werte der Features in stark unterschiedlichen Bereichen liegen.
